```python
# import libraries
from pandas import DataFrame, read_csv, to_datetime, merge
import matplotlib.pyplot as plt
```

```python
# Set inline plotting
%matplotlib inline
```

```python
# File location
f = r'C:\Users\david\Downloads\transactions.csv'

# csv file is tab seperated
# make the Date column the index of the dataframe
# ask Pandas to identify date columns 
raw = read_csv(f, sep = '\t', index_col = 'Date', parse_dates = True)
#raw
```

Let us take a peek at the data we will be working with. Note that I had to do a small amount of work in cleaning the data before I could begin my analysis.    

**Steps I took to clean the data:**  

* Sort the Date column ascending   
* Make the Amount column positive for credits and negative for debits  

```python
# we create a new dataframe with the date column ordered
df = raw.sort_index()
#df.head()
```

```python
# debit = -1 & credit = 1
AddSigns = df['Transaction Type'].apply(lambda x: -1 if x == 'debit' else 1).values

# Add signs to Amount column
df['Amount'] = df['Amount']*AddSigns

df.head(15)
```

Now that the data is clean, we can begin our analysis.  

We simply start out by plotting the only numerical column, Amount! It seems like there is a steady positive cashflow thoughout the year. On the debit side, the winter and summer months seem to have slightly more activity than the other months. April, June, September, and November are months we can keep an eye out for. 

```python
df['Amount'].plot();
```

Most of the transactions seem to be between 5 and 28 dollars. Other figures to keep an eye out for are 40, 1,343.74 (credit), and 694.

As you can see, we are on the money a good chunck of the time. Lets try to see if we can find the missing atm withdrawls.

```python
# since there are multiple transaction per day, we can resample the data so
# that transactions that happen on the same day are aggregated.
dailyatm = df[atm].resample(rule='d', how='sum')

#Create a figure
fig, axes = plt.subplots(2,1,figsize=(12,5))

# plot
df['Amount'][FortyDollars].plot(kind='bar', color='r', ax = axes[0], title = 'FortyDollars');
dailyatm['Amount'].dropna().plot(kind='bar',ax = axes[1], title = 'ATM Fees');

# Adjust plots
plt.subplots_adjust(wspace=0, hspace=2.5)
```

```python
# get all ATM withdrawals
withdrawals = df['Description'].apply(lambda x: x.lower()).str.contains('withdraw')

# problem is this is also picking up atm fees, we fix that here
#df[withdrawals & (df['Category'] <> 'ATM Fee')]
```

Things are getting a little better for us, we are definitely seeing a pattern here. Although it does seem that not every withdrawal ties to an ATM fee. It is also not clear from the graph, so lets try a little harder... 

```python
#Create a figure
fig, axes = plt.subplots(2,1,figsize=(12,5))

# plot
df['Amount'][withdrawals & ~(df['Description'] == 'Withdrawal') & (df['Category'] <> 'ATM Fee')].plot(kind='bar', color='r',ax = axes[0]);
dailyatm['Amount'].dropna().plot(kind='bar',ax = axes[1]);

# adjust plots
plt.subplots_adjust(wspace=0, hspace=1.7)
```

```python
# We make dataframes out of the two slices of data
w = df['Amount'][withdrawals & ~(df['Description'] == 'Withdrawal') & (df['Category'] <> 'ATM Fee')]
f = dailyatm['Amount'].dropna()

w = DataFrame(w)
f = DataFrame(f)

# We combine the data sets by their index
combined = merge(w,f, left_index=True, right_index=True, how='outer', suffixes=('_Withdrawals', '_Fees'))
```

This is soo much easier to understand. We can now clearly see which were the days that we saw both fees and withdrawls. 

**Possible reasons for the missing fees:**  

* Customer used an ATM that does not charge them a fee  
* The category "ATM Fee" is not complete. There might be atm fees that were not tagged as such.  


```python
combined.plot(kind='bar',figsize=(12,5));
```
